### Built-in Exceptions
The simplest way to handle exceptions is with a "try-except" block:


```python
#example 1: ZeroDivisionError
try:
    # Some code that might be raising zero division error.
    x = 1
    print( x / 0 )
except ZeroDivisionError as e:
    print("You are trying to divide by zero!")
    
    print(x / 1)
```

```python
#example 2: ValueError

#If user enters a string
x = int(input("Please enter a number: "))
```

```python
#example 3: except ZeroDivisionError and except ValueError
try:
    
    x = int(input("Please enter a number: "))
    print( 1 / x )
except ZeroDivisionError:
    print("You are trying to divide into zero! Start over and enter different number than zero.")
except ValueError:
    print("Oops!  That was no valid number.  Try again...")
```

```python
#example 4: except(ValueError, ZeroDivisionError)
try:
    
    x = int(input("Please enter a number: "))
    print( 1 / x )
except (ValueError, ZeroDivisionError):
    print("This is not a valid input. Please enter a number between [1,N].")
```

```python
# example 5:try...except...as...
try:
    
    x = int(input("Please enter a number: "))
    print( 1 / x )
except Exception as e:
    print("Something occured: ", e)
```

```python
#example 6:try...except...else...
try:
    # Hata olma olasl olan bir kod.
    x = int(input("Please enter a number: "))
    print( 1 / x )
except Exception as e:
    print(e)
else:
    print("The user seems watchful, there is no exception!")
```

```python
#example 7:try...except...finally...
try:
    
    x = int(input("Please enter a number: "))
    print( 1 / x )
except Exception as e:
    print(e)
finally:
    print("This operation is required either for exception cases or normal cases.")
```

```python
try:
    import seaborn

except Exception as e:
    print("An error occured", e)
```

### Some Examples using Exceptions

- **except IOError**:<br>
    >print('An error occured trying to read the file.')<br>
- **except ValueError**:<br>
    >print('Non-numeric data found in the file.')<br>
- **except ImportError**:<br>
    > print ("NO module found)<br>
- **except EOFError**:<br>
    > print('Why did you do an EOF on me?')<br>
- **except KeyboardInterrupt**:<br>
    > print('You cancelled the operation.')<br>
- **except**:<br>

```python
try:
    file = open("myfileblabla.txt", "r")
except IOError:
    print("IOError: File can not be opened.")
```

```python
try:
    data = int(input("Do you want to continue?: "))
    
    print(data * 10)
    
except ValueError:
    print("ValueError: You should enter an integer.")
    data = ""
    print(data)
```

```python
try:
    import seaborn
except ImportError as e:
    print("ImportError: Can not import some library, ", e)
```

```python
try:
    text = input('Enter something --> ')
except EOFError:
    print('Why did you do an EOF on me?')
except KeyboardInterrupt:
    print('You cancelled the operation.')
else:
    print('You entered {}'.format(text))
```

### User-Defined Exceptions

```python
#example 11: User-defined Exceptions

# A python program to create user-defined exception
 
# class MyError is derived from super class Exception
class MyError(Exception):
 
    # Constructor or Initializer
    def __init__(self, value):
        self.value = value
 
    # __str__ is to print() the value
    def __str__(self):
        return(repr(self.value))
 

try:
    raise(MyError(3*2))
 
# Value of Exception is stored in error
except MyError as error:
    print('A New Exception occured: ',error.value)
    
    print(error.value / 2)
```

```python
#example 12: User-defined exceptions

# class Error is derived from super class Exception
class Error(Exception):
 
    # Error is derived class for Exception, but
    # Base class for exceptions in this module
    pass
 
class TransitionError(Error):
 
    # Raised when an operation attempts a state
    # transition that's not allowed.
    def __init__(self, prev, nex, msg):
        self.prev = prev
        self.next = nex
 
        # Error message thrown is saved in msg
        self.msg = msg
try:
    raise(TransitionError(2,3*2,"Not Allowed"))
 
# Value of Exception is stored in error
except TransitionError as error:
    print('Exception occured: ',error.msg, error)
```
