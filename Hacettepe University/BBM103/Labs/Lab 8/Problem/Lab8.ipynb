### <font color=red>WHAT IS RECURSION? </font> 
* **Goal:** simplify the problem by solving the same problem for smaller input
> * Solve problems by **divide(decrease)-and-conquer**
* Function calls itself (but not infinitely!) 
> * One or more base cases

### <font color=red>ITERATION vs. RECURSION  </font> 
* An **ITERATIVE** function is one that loops to repeat some part of the code.<br>
* A **RECURSIVE** function is one that calls itself again to repeat the code.

### <font color=red>ADVANTAGES and DISADVANTAGES OF RECURSION    </font>
* Recursion may be simpler, more intuitive, and also efficient and natural for a programmer.<br>
* BUT! Recursion may not be efficient from the computerspoint of view.<br> 
> * Ex. Computing nth Fibonacci number recursively takes O(2**n) steps!
